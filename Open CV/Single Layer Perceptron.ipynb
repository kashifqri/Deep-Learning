```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
datasets = pd.read_csv("C://Users//basit//Downloads//Placement (1).csv")
```

```python
datasets
```

```python
sns.scatterplot(x="CGPA",y="IQ",hue="Placement",data=datasets)
plt.show()
```

```python
x=datasets.iloc[:,:-1]
y= datasets["Placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import Perceptron
```

```python
pr =Perceptron()
pr.fit(x_train,y_train)
```

```python
pr.score(x_train,y_train)*100 , pr.score(x_test,y_test)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=pr)
plt.show()
```
